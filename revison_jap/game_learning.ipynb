```python
import pandas as pd
import pickle
import random
import customtkinter
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import tkinter
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("nihongoN5List.pkl", "rb") as f:
    dfs = pickle.load(f)
```

## flowchart

- start
- after
- datamang
- testing
- checking ----- can go back to testing
- score ---- can go back to start

```python
customtkinter.set_appearance_mode("dark")
customtkinter.set_default_color_theme("dark-blue")


root = customtkinter.CTk()
root.geometry("500x350")
root.title("Japanese N5 Learning")
counter = 0
total = 0
dicts = {}
lists = []


for i in dfs:
    lens = len(i)

    number = len(i) // 10

    if number == 0:
        lists.append(["All"])

    elif lens / 10 > int(lens / 10):
        res = list(range(10, len(i), 10))
        res = [str(num) for num in res]
        res.append("All")
        lists.append(res)
    else:
        res = list(range(10, len(i) + 10, 10))
        res = [str(num) for num in res]
        lists.append(res)
list1 = [
    "あ",
    "い",
    "う",
    "え",
    "お",
    "か",
    "き",
    "く",
    "け",
    "こ",
    "さ",
    "し",
    "す",
    "せ",
    "そ",
    "た",
    "ち",
    "つ",
    "て",
    "と",
    "な",
    "に",
    "ぬ",
    "ね",
    "の",
    "は",
    "ひ",
    "ふ",
    "へ",
    "ほ",
    "ま",
    "み",
    "む",
    "め",
    "も",
    "や",
    "よ",
    "ら",
    "り",
    "れ",
    "ろ",
    "わ",
]
category = {key: value for key, value in zip(list1, lists)}

frame = customtkinter.CTkScrollableFrame(master=root)
frame.pack(pady=20, padx=60, fill="both", expand=True)


def start(dicts):

    clear_frame(frame)
    label = customtkinter.CTkLabel(master=frame, text="Select the Data")
    label.pack(pady=12, padx=10)

    combobox = customtkinter.CTkComboBox(master=frame, state="readonly", values=list1)
    combobox.pack(padx=20, pady=10)
    combobox.set("あ")

    cont = customtkinter.CTkButton(
        master=frame,
        text="continue",
        command=lambda: after(combobox, dicts),
    )
    cont.pack(pady=12, padx=10)


def after(combobox, dicts):
    alpha = combobox.get()

    clear_frame(frame)

    label = customtkinter.CTkLabel(master=frame, text="Number of words to focus on :")
    label.pack(pady=12, padx=10)
    combobox2 = customtkinter.CTkComboBox(
        master=frame, state="readonly", values=category[alpha]
    )
    combobox2.pack(padx=20, pady=10)
    combobox2.set(category[alpha][0])

    button = customtkinter.CTkButton(
        master=frame,
        text="continue",
        command=lambda: data_manag(alpha, combobox2, dicts),
    )
    button.pack(pady=12, padx=10)
    back = customtkinter.CTkButton(
        master=frame,
        text="back",
        command=lambda: start(dicts),
    )
    back.pack(pady=12, padx=10)


def score(counter, total, frame, dicts):

    clear_frame(frame)
    Final = customtkinter.CTkLabel(
        master=frame, text=f"Your Score is ----{counter}/{total}----"
    )
    Final.pack(pady=12, padx=10)
    percent = round((counter / total) * 100, 2)
    Final2 = customtkinter.CTkLabel(
        master=frame, text=f"Your percentage is ----{percent}%----"
    )
    Final2.pack(pady=12, padx=10)

    if total == counter:
        perfect = customtkinter.CTkLabel(
            master=frame,
            text=f"-------- CONGRATULATION ON A PERFECT SCORE --------",
            text_color="green",
        )
        perfect.pack(pady=12, padx=10)

    else:
        Final4 = customtkinter.CTkLabel(
            master=frame, text="-------- Your Mistakes --------"
        )
        Final4.pack(pady=12, padx=10)
        keys = list(dicts.keys())
        values = list(dicts.values())
        fig = Figure(figsize=(20, 10), dpi=100)
        ax = fig.add_subplot(111)
        ax.barh(keys, values, color="blue")
        ax.set_yticks(np.arange(len(keys)))
        ax.set_yticklabels(keys)
        canvas = FigureCanvasTkAgg(fig, master=frame)
        canvas.draw()
        canvas.get_tk_widget().pack(side=tkinter.TOP)

    menu = customtkinter.CTkButton(
        master=frame, text="Menu", command=lambda: start(dicts)
    )
    menu.pack(pady=12, padx=10)


def data_manag(alpha, combobox2, dicts):

    df = dfs[list1.index(alpha)]
    if combobox2.get() == "All":
        lenght = len(df) - 1
    else:
        lenght = int(combobox2.get()) - 1
    dicts = {}
    testing(df, lenght, counter, total, dicts)


def clear_frame(frame):
    for widget in frame.winfo_children():
        widget.destroy()


def checking(answer, isit, frame, df, lenght, counter, total, dicts, meaning, button2):
    button2.destroy()
    if answer == isit:
        ok = customtkinter.CTkLabel(master=frame, text="Correct", text_color="green")
        ok.pack(pady=12, padx=10)
        counter = counter + 1

    else:
        ok = customtkinter.CTkLabel(
            master=frame, text=f"Wrong: answer is ----{answer}----", text_color="red"
        )
        ok.pack(pady=12, padx=10)
        if len(dicts) == 0:
            dicts[str(meaning + " , " + answer)] = 1
        elif str(meaning + " , " + answer) in dicts:
            dicts[str(meaning + " , " + answer)] = (
                dicts[str(meaning + " , " + answer)] + 1
            )
        elif str(meaning + " , " + answer) not in dicts:
            dicts[str(meaning + " , " + answer)] = 1
    total = total + 1

    button3 = customtkinter.CTkButton(
        master=frame,
        text="Continue",
        command=lambda: testing(df, lenght, counter, total, dicts),
    )
    button3.pack(pady=12, padx=10)
    exit = customtkinter.CTkButton(
        master=frame, text="EXIT", command=lambda: score(counter, total, frame, dicts)
    )
    exit.pack(pady=12, padx=10)


def testing(df, lenght, counter, total, dicts):
    clear_frame(frame)

    if lenght == 0:
        warning = customtkinter.CTkLabel(
            master=frame, text=f"The length of this data is 1", text_color="red"
        )
        warning.pack(pady=12, padx=10)
    num = random.randint(0, lenght)
    meaning = df.iloc[num]["Meaning"]
    mean = customtkinter.CTkLabel(
        master=frame, text=f"Find the meaning of ----{meaning}----"
    )
    mean.pack(pady=12, padx=10)
    isit = customtkinter.CTkEntry(master=frame)
    isit.pack(pady=12, padx=10)
    button2 = customtkinter.CTkButton(
        master=frame,
        text="Check",
        command=lambda: checking(
            df.iloc[num]["Romaji"],
            isit.get(),
            frame,
            df,
            lenght,
            counter,
            total,
            dicts,
            meaning,
            button2,
        ),
    )
    button2.pack(pady=12, padx=10)


start(dicts)
root.mainloop()
```
